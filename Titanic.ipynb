```python
import pandas as pd
import sqlite3
#Analize it and also write it into a datase
```

```python
titanic = "datasets/titanic.xls"
df = pd.read_excel(titanic)
```

```python
# get the number of persons per class
ppc = df.groupby("pclass")["name"].count()
ppc
```

```python
# get the number of persons tht servived per class
spc = df.groupby("pclass")["survived"].sum()
spc
```

```python
survival_rate = round(spc / ppc * 100, 1)
print(survival_rate)
srdf = pd.DataFrame({"persons per class":ppc, "survived per class":spc, "survival rate per class (%)": survival_rate})
print(srdf)
```

```python
median_age = df["age"].median()
mean_age = df["age"].mean()
median_age_to_survive = df.groupby("survived")["age"].median()
mean_age_to_survive = df.groupby("survived")["age"].mean()
print("age on the boat:")
print(mean_age)
print(median_age)
print("age to survive:")
print(mean_age_to_survive)
print(median_age_to_survive)

# age does not matter
```

```python
max_ticket_price = df["fare"].max()
min_ticket_price = df["fare"].min()
print("cheapest ticket:", min_ticket_price)
print("most expensive ticket:", max_ticket_price)

median_tickt_price=df.groupby("pclass")["fare"].median()
mean_tickt_price=df.groupby("pclass")["fare"].mean()
print("avarage cost of a ticket in each class: (median):")
print(median_tickt_price)
print("avarage cost of a ticket in each class: (mean)")
print(mean_tickt_price)
```

```python
mean_price_to_survive=df.groupby("survived")["fare"].mean()
median_price_to_survive=df.groupby("survived")["fare"].median()
print("what you pay to survive: (median)")
print(median_price_to_survive)
print("what you pay to survive: (mean)")
print(mean_price_to_survive)
```

```python
# age in class
median_age_in_class = df.groupby("pclass")["age"].median()
mean_age_in_class = df.groupby("pclass")["age"].mean()
print(median_age_in_class)
print(mean_age_in_class)
```

```python
# oldest person on board:
oldest = df["age"].max()
# youngest person on board:
youngest = df["age"].min()
# average age on board:
mean = df["age"].mean()
median = df["age"].median()
std = df["age"].std()
print("oldest:", oldest, "youngest:", youngest, "  - avg age (mean, median):", mean, median, std)
```

```python
conn = sqlite3.connect(':memory:')
c = conn.cursor()
```

```python
c.execute("""CREATE TABLE titanic (
            class integer,
            survived boolean,
            name text,
            sex text,
            age float,
            fare float
            )""")
```

```python
# c.execute("""insert into titanic (class, survived, name, sex, age, fare) VALUES (?, ?, ?, ?, ?, ?)""",
#          (list(df["pclass"]), df["survived"], df["name"], df["sex"], df['age'], df['fare']))


c.execute("delete from titanic")
cls = list(df["pclass"])
svd = list(df["survived"])
nam = list(df["name"])
sex = list(df["sex"])
age = list(df["age"])
far = list(df["fare"])

for i in range(len(cls)):
    c.execute("""insert into titanic (class, survived, name, sex, age, fare) VALUES (?, ?, ?, ?, ?, ?)""",
        (cls[i], svd[i], nam[i], sex[i], age[i], far[i]))
    
conn.commit()
c.execute("select * from titanic").fetchall()
```

```python
c.execute("delete from titanic")

for i in range(len(df)):
    c.execute("""insert into titanic (class, survived, name, sex, age, fare) VALUES (?, ?, ?, ?, ?, ?)""",
        (int(df["pclass"][i]), bool(df["survived"][i]), df["name"][i], df["sex"][i], df["age"][i], df["fare"][i]))
    
conn.commit()
c.execute("select * from titanic").fetchall()
```
